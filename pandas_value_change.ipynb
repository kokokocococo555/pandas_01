## Pythonのpandasデータフレームの要素の値が変更される場合と変更されない場合

### # case1: 代入が反映されない場合
- 行を指定->列を指定　はダメ

- 仮説1: バージョンによる？
    - いけた環境
        - Windows10
        - Python3.6.7
        - pandas0.24.2
    - いけなかった環境
        - Ubuntu16.04 LTS
        - Python3.6.7
        - pandas0.24.2
    - 特に変わらず

- 仮説2: 元の値の型による？
    - 以下の実験の通り、元が数値なら上書きされてnp.nanだと上書きできない

```python
# 代入が反映されない
i = 0
c = "a"
df.iloc[i].loc[c] = 111
df.head()
```

```python
# 元が数値だと代入が反映される様子
i = 0
c = "a"
df_0.iloc[i].loc[c] = 111
df_0.head()
```

### # case2: 代入が反映される場合
- 列(Series)を指定->行を指定（スライシング）　はOK

#### 2-1: エラーが出る

```python
# これはエラーが出る
i = 0
c = "a"
df.loc[c][i] = 211
df.head()
```

#### 2-2: 列指定からの行指定

```python
# これならOK
# 元がnp.nanだとWarningが出る
# 元が数値だとWarningは出ない（以下同様）
i = 0
c = "a"
df[c][i] = 221
df.head()
```

```python
# こちらもOK
i = 0
c = "a"
df.loc[i, c] = 222
df.head()
```

```python
# atを使った方が速いらしい
# https://note.nkmk.me/python-pandas-at-iat-loc-iloc/
i = 0
c = "a"
df.at[i, c] = 223
df.head()
```

```python
# 組み合わせ技
# Warningが出ることがある
i = 0
c = "b"
df[c].iloc[i] = 224
df.head()
```

```python
# df.index[i]でインデックスを取得できる
i = 0
c = "a"
df.loc[df.index[i], c] = 225
df.head()
```

## 指定しているのは行番号かインデックスか

```python
# この代入の場合、元のデータも変更される
df2 = df[::-1]
```

```python
df2.head()
```

```python
# dfの行番号ではなく、indexを指定している
i = 0
c = "a"
df2[c][i] = 231
df2.head()
```

```python
# 元のデータも変更されている
df.head()
```

```python
# これも同様にdfの行番号ではなく、indexを指定している
i = 0
c = "a"
df2.loc[i, c] = 232
df2.head()
```

```python
# atを使っても同様
i = 0
c = "a"
df2.at[i, c] = 233
df2.head()
```

```python
# .ilocはインデックスではなく行の順序で指定している様子
i = 0
c = "a"
df2[c].iloc[i] = 234
df2.head()
```

```python
# df.index[i]はインデックスをとってくる様子
i = 0
c = "a"
df2.loc[df.index[i], c] = 235
df2.head()
```

## 結論

### dfの値の変更時は 「列(Series)を指定->行を指定（スライシング）」
- pd.DataFrameはpd.Seriesが集まったものだから、最初にSeriesを取り出す必要がある？

### 行名・列名で指定するなら`.at[i, c]`
- `df.at[i, c] = 223`

#### これは列の指定でエラーが出る
- `df.loc[c][i] = 210`

### 行の順番・列名で指定するなら`.iloc[i]`
- `df[c].iloc[i] = 999`
